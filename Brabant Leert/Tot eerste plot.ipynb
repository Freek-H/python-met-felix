```python
from datetime import datetime
def calculate_age(birth_date):
    current_date = datetime.now()
    if current_date.month > birth_date.month:
        return current_date.year - birth_date.year
    elif current_date.month == birth_date.month and current_date.day >= birth_date.day:
        return current_date.year - birth_date.year
    else:
        return current_date.year - birth_date.year - 1
```

```python
df_age = df.copy()
df_age['age'] = [calculate_age(birth_date) for birth_date in df_age['birth_date']]
```

```python
plt.scatter(df_age.age, df_age.salary, alpha=.1)
```

```python
ages = sorted(df_age['age'].unique())
ages
```

```python
avg_salary_per_age = []
for one_age in ages:
    avg_salary_per_age.append(df_age['salary'][df_age['age'] == one_age].mean())
```

```python
import numpy as np
```

```python
plt.plot([age - 7 for age in ages], avg_salary_per_age)
plt.plot([age - 7 for age in ages], np.array(avg_salary_per_age) - np.array([df_age['salary'][df_age['age'] == one_age].std() for one_age in ages]), color='black')
plt.plot([age - 7 for age in ages], np.array(avg_salary_per_age) + np.array([df_age['salary'][df_age['age'] == one_age].std() for one_age in ages]), color='black')
```

```python
ranges = []
lower_ranges = []
for one_range in df['commute_distance']:
    one_range = one_range.rstrip(' Miles')
    if one_range[-1] == '+':
        start_range = int(one_range.rstrip('+'))
        end_range = None
    else:
        start_range, end_range = one_range.split('-')
        start_range = int(start_range)
        end_range = int(end_range)
    ranges.append((start_range, end_range))
    lower_ranges.append(start_range)
lower_ranges
```

```python
df_commute = df_age.copy()
df_commute['commute_distance'] = lower_ranges
```
