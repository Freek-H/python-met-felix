```python
replacements = {
    '{titel}': 'Stap twee',
    '{header}': 'Stap twee',
    '{text}': 'We doen hier stap twee, zie de link voor een tabel',
    '{link text}': 'Open de tabel',
    '{link}': '"tabel.html"'
}

with open('template.html', 'r') as input_file:
    with open('index3.html', 'w') as output_file:
        html = input_file.read()
        for key, value in replacements.items():
            html = html.replace(key, value)
        output_file.write(html)
```
